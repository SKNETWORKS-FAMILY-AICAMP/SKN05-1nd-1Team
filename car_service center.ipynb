# 서비스 센터

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time  # sleep(1)  1초
from selenium.webdriver.common.by import By
import pandas as pd
from tqdm import tqdm
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

### 현대

```python


# 지역검색
wd = webdriver.Chrome()
wd.get('https://www.hyundai.com/kr/ko/service-membership/service-network/service-reservation-search') # 크롤링할 주소를 입력
html = wd.page_source  # 페이지 정보를 가지고 있음  html정보

soup = BeautifulSoup(html, 'html.parser') # html 정보를 파싱

dropdown = wd.find_element(By.XPATH,'/html/body/div[2]/div[2]/dl/dl/dd[2]/button[2]')
dropdown.click()
time.sleep(0.5)   

result = []
dropdown = wd.find_element(By.XPATH,'//*[@id="selectBoxTownShip"]')
dropdown.click()
time.sleep(0.5)
dropdown = wd.find_element(By.XPATH,'//*[@id="selectBoxTownShip"]/option[1]')
dropdown.click()

time.sleep(0.5)
for i in range(2,19):
    wd.execute_script("window.scrollTo(0, 0);")  # 페이지 최상단으로 이동
    
    dropdown = wd.find_element(By.XPATH,'//*[@id="selectBoxCity"]')
    dropdown.click()
    time.sleep(0.5)   
        
    dropdown = wd.find_element(By.XPATH,f'//*[@id="selectBoxCity"]/option[{i}]')
    dropdown.click()    
    time.sleep(0.5)  
            
    dropdown = wd.find_element(By.XPATH,'//*[@id="contents"]/div[2]/div/div[3]/div[1]/div/div[2]/a[2]')
    dropdown.click()
    time.sleep(0.5)
    while(True): # 페이지가 없을때까지 반복
        html = wd.page_source # 현재 페이지 정보를 html로 변환
        soup = BeautifulSoup(html, 'html.parser')
        shop_lists = soup.select('tbody#listView > tr')
        # 각 서비스센터 정보를 result 리스트에 저장
        for shop in shop_lists:
            # 테이블 행에서 각 열(td) 가져오기
            tds = shop.find_all('td')
            if len(tds) >= 3:  # 데이터가 있는지 확인
                center_name = tds[0].text.strip()  # 서비스센터 이름.
                address = tds[1].text.strip()  # 주소
                phone = tds[2].find('span','pc_only').text.strip()  # 전화번호
                result.append([center_name, address, phone])
        # 다음 페이지가 없으면 break, 다음 페이지가 있으면 다음 페이지로 이동
        next_button = wd.find_element(By.CSS_SELECTOR, '#pageNavi > button.navi.next')  # 다음 페이지 버튼 선택자
        if 'navi next disabled' in next_button.get_attribute('class'):  # 버튼이 비활성화된 경우 종료
            break
        next_button.click()
        time.sleep(0.5)

df = pd.DataFrame(result)
# DataFrame을 CSV 파일로 저장
df.to_csv('hyundai_service_centers.csv', index=False, encoding='utf-8-sig')  # UTF-8로 인코딩
```

### 르노

```python
# 웹 크롤링 시작: 르노 서비스 센터 페이지 열기
wd = webdriver.Chrome()
wd.get('https://www.renault.co.kr/ko/service/svc_maintenance.jsp')  # 크롤링할 URL 입력
time.sleep(2)  # 페이지 로딩 대기

result = []  # 결과 저장할 리스트 초기화

# 모델을 선택하는 루프
for model_idx in range(2, 5):  # 2~4번 모델 선택 (예시 범위)
    try:
        dropdown_model = wd.find_element(By.XPATH, '//*[@id="modelselect"]')  # 모델 선택 드롭다운 클릭
        dropdown_model.click()
        time.sleep(0.1)  # 잠시 대기
    
        model_option = wd.find_element(By.XPATH, f'//*[@id="modelselect"]/option[{model_idx}]')  # 각 모델 선택
        model_option.click()    
        time.sleep(0.1)
    
        # 지역을 선택하는 루프
        for region_idx in range(2, 19):  # 2~18번 지역 선택 (서울특별시 ~ 제주도)
            dropdown_region = wd.find_element(By.XPATH, '//*[@id="netsido"]')  # 지역 선택 드롭다운 클릭
            dropdown_region.click()
            time.sleep(0.1)
    
            region_option = wd.find_element(By.XPATH, f'//*[@id="netsido"]/option[{region_idx}]')  # 각 지역 선택
            region_option.click()    
            time.sleep(0.1)
    
            # 현재 페이지 소스를 HTML로 변환하여 BeautifulSoup으로 파싱
            html = wd.page_source  
            soup = BeautifulSoup(html, 'html.parser')
    
            # 서비스 센터 리스트를 포함하는 div 요소 선택
            
            shop_lists = soup.select('#store_list > div')
            for sh in shop_lists:
                tds=sh.find_all('div', class_="info") 
                result.append( [tds[0].strong.text,  tds[0].select('li')[0].text,  tds[0].select('li')[1].text] )
    except Exception as e:
        print(f'error model_idx : {model_idx}')
        print(e)
        #오류났을 때 몇번째에서 오류가 났는지

df = pd.DataFrame(result, columns=["센터 이름", "주소", "전화번호"])

# 중복된 행 제거
df_cleaned = df.drop_duplicates()


print("\n중복 제거 후 DataFrame:")
print(df_cleaned)

# DataFrame을 CSV 파일로 저장
df.to_csv('Renault_Korea_service_centers.csv', index=False, encoding='utf-8-sig')  # UTF-8로 인코딩
```

### 테슬라

```python

# 웹 크롤링 시작: 테슬라 서비스 센터 페이지 열기
wd = webdriver.Chrome()
wd.get('https://www.tesla.com/ko_KR/findus/list/services/South%20Korea')  # 크롤링할 URL 입력
time.sleep(5)  # 페이지 로딩 대기 (동적 로딩을 고려하여 시간 늘림)

result = []  # 결과 저장할 리스트 초기화

# 현재 페이지 소스를 HTML로 변환하여 BeautifulSoup으로 파싱
html = wd.page_source  
soup = BeautifulSoup(html, 'html.parser')

# 서비스 센터 리스트를 포함하는 div 요소 선택
shop_lists = soup.select('#find-us-list-container > div > section > div')

for sh in shop_lists:
    try:
        # 각 서비스 센터 정보를 포함하는 address 요소 선택
        addresses = sh.find_all('address', class_='vcard')
        for address in addresses:
            center_name = address.find('div', class_='anchor-container').a.text.strip() if address.find('div', class_='anchor-container') and address.find('div', class_='anchor-container').a else '정보 없음'
            
            street_address = address.find('span', class_='street-address-states').text.strip() if address.find('span', class_='street-address-states') else '주소 없음'
            extended_address = address.find('span', class_='extended-address-states').text.strip() if address.find('span', class_='extended-address-states') else ''
            full_address = f"{street_address} {extended_address}".strip()
            
            phone_number = address.find('span', class_='value').text.strip() if address.find('span', class_='value') else '전화번호 없음'
            
            # 결과 리스트에 추가
            result.append([center_name, full_address, phone_number])
    except Exception as e:
        # 예외 발생 시 오류 메시지 출력
        print(f'Error processing shop: {sh}')
        print(e)

# DataFrame으로 변환 및 중복 제거
df = pd.DataFrame(result, columns=['센터 이름', '주소', '전화번호'])
df.drop_duplicates(inplace=True)

# DataFrame을 CSV 파일로 저장
df.to_csv('tesla_service_centers.csv', index=False, encoding='utf-8')  # UTF-8로 인코딩
print(df)

# 브라우저 종료
wd.quit()
```

### BMW

```python

# 웹 크롤링 시작: BMW 서비스 센터 페이지 열기
wd = webdriver.Chrome()
wd.get('https://www.bmw.co.kr/ko/fastlane/dealer-locator/service-center.html')  # 크롤링할 URL 입력
time.sleep(2)  # 페이지 로딩 대기

# iframe 찾기
iframe = wd.find_element(By.TAG_NAME, 'iframe')

# iframe으로 전환
wd.switch_to.frame(iframe)

# iframe 내의 HTML 추출하기
html_content = wd.page_source

# BeautifulSoup으로 HTML 파싱하기
soup = BeautifulSoup(html_content, 'html.parser')
li_elements = soup.find_all('li', class_='content_area_choise_item')
result = []

for li in li_elements:
    try:
        # 서비스 센터 이름 추출
        name_element = li.find('a', class_='content_area_choise_item_anchor')
        name = name_element.text.strip() if name_element else 'N/A'
        
        # 주소 정보 추출
        info_desc = li.find('div', class_='content_area_choise_info_desc')
        address_doro = info_desc['data-address-doro'] if info_desc and 'data-address-doro' in info_desc.attrs else 'N/A'
        address_jibun = info_desc['data-address-jibun'] if info_desc and 'data-address-jibun' in info_desc.attrs else 'N/A'
        
        # 전화번호 추출
        phone_div = info_desc.find('div', class_='content_area_choise_info_desc_left')
        phone_element = phone_div.find('div', class_='content_area_choise_info_desc_item first') if phone_div else None
        phone = phone_element.get_text(strip=True).replace('전화번호', '').strip() if phone_element else 'N/A'
        
        # 영업시간 추출
        hours_div = info_desc.find('div', class_='content_area_choise_info_desc_right')
        hours_element = hours_div.find('div', class_='content_area_choise_info_desc_item') if hours_div else None
        hours = hours_element.get_text(strip=True).replace('영업시간', '').strip() if hours_element else 'N/A'
        
        # 웹사이트 추출
        website_element = info_desc.find('a', class_='jq_dealer_link')
        website = website_element['href'] if website_element else 'N/A'
        
        result.append([name, address_doro, address_jibun, phone, hours, website])
        
    except Exception as e:
        print(e)

# DataFrame으로 변환 및 중복 제거
df = pd.DataFrame(result, columns=['센터 이름', '주소 (도로명)', '주소 (지번)', '전화번호', '영업시간', '웹사이트'])
df.drop_duplicates(inplace=True)

# DataFrame을 CSV 파일로 저장
df.to_csv('bmw_service_centers.csv', index=False, encoding='utf-8')  # UTF-8로 인코딩
print(df)
```
